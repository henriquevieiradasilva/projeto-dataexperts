### Parte 2 - **Exclusão do Data Lakehouse**

`Insira na variável nome_datalakehouse o nome do lakehouse que deseja excluir:`

```python
nome_datalakehouse = "dataexperts"
```

Aqui será criado uma **variável** para ajudar a **validar** se o Data Lakehouse **já existe**:

```python
datalakehouse_existe = True
```

Aqui é o código para buscar todos os Data Lakehouses existentes e armazenar o resultado dessa consulta em um Data Frame, assim verificando se o nome que estamos procurando já existe
  - Se não existir, não acontecerá mais nada.
  - Se já existir, exclui o Data LakeHouse e todas as suas respectivas camadas

(tudo isso é possível usando a variável ***datalakehouse_existe***) 

O código a seguir descobre se existe ou não:

```python
dataslakeshouses_existentes = spark.sql("SHOW CATALOGS;")

datalakehouse_existe = (
    dataslakeshouses_existentes
    .filter(dataslakeshouses_existentes.catalog == nome_datalakehouse)
    .count() > 0
)
```

O código a seguir exclui se existir:

```python
if datalakehouse_existe:
    spark.sql(f"DROP CATALOG {nome_datalakehouse} CASCADE;")
```

### **Resultado Final**

Aqui é um log simples para mostrar se deu certo ou não toda a construção da arquitetura planejada.


```python
# Código está comentado por ser apenas uma demonstração. Para visualizar o resultado, retire o identificador de comentário e execute novamente

if datalakehouse_existe:
    print(f"[LOG] O Data LakeHouse '{nome_datalakehouse}' foi excluido com sucesso!")
else:
    print(f"[LOG] O Data LakeHouse '{nome_datalakehouse}' não existe ainda, nenhuma alteração foi feita!")
```
